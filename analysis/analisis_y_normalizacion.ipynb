```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Cargamos el dataset etiquetado
df = pd.read_csv('dataset_etiquetado.csv')

# Mostramos las primeras filas
df.head()
```

```python
# Conteo de muestras por gesto
conteo_gestos = df['gesture'].value_counts()

# Mostrar el conteo
print(conteo_gestos)

# Visualización del conteo
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='gesture', order=conteo_gestos.index, palette='viridis')
plt.title('Cantidad de muestras por gesto')
plt.xticks(rotation=45)
plt.show()
```

```python
# Seleccionamos solo las columnas numéricas (sin 'label' ni 'gesture')
df_numerico = df.drop(['label', 'gesture'], axis=1)

# Calculamos la matriz de correlación de Spearman
corr_spearman = df_numerico.corr(method='spearman')

# Mostramos una porción de la matriz
corr_spearman.head()
```

```python
# Mapa de calor completo
plt.figure(figsize=(18,14))
sns.heatmap(corr_spearman, cmap='coolwarm', square=True, cbar=True)
plt.title('Mapa de calor de correlación (Spearman)')
plt.show()
```

```python
# Seleccionamos algunas features relevantes para visualizar
features_vis = ['standard_deviation_e1', 'root_mean_square_e1', 'minimum_e1', 'maximum_e1']

# Pairplot con Seaborn
sns.pairplot(df, vars=features_vis, hue='gesture', palette='muted')
plt.suptitle('Agrupación de gestos en espacio de características seleccionadas', y=1.02)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Creamos un scaler
scaler = MinMaxScaler()

# Seleccionamos las columnas numéricas de features
features_cols = df.drop(['label', 'gesture'], axis=1).columns

# Normalizamos las features
df_normalized_features = pd.DataFrame(scaler.fit_transform(df[features_cols]), columns=features_cols)

# Reagregamos las columnas 'label' y 'gesture'
df_normalized = pd.concat([df_normalized_features, df[['label', 'gesture']]], axis=1)

# Mostramos las primeras filas normalizadas
df_normalized.head()
```

```python
# Guardamos el dataset normalizado en un nuevo CSV
df_normalized.to_csv('emg_normalized_labeled.csv', index=False)

print("✅ Dataset normalizado y guardado como 'emg_normalized_labeled.csv'")
```

```python
# Cargar el dataset normalizado
df_normalized = pd.read_csv('emg_normalized_labeled.csv')

# Verificar
df_normalized.head()
```

```python
# Seleccionamos solo las features (sin 'label' ni 'gesture')
features_cols = df_normalized.drop(['label', 'gesture'], axis=1)

# Calculamos la correlación de Spearman
corr_spearman_normalized = features_cols.corr(method='spearman')

# Mostramos una parte de la matriz de correlación
corr_spearman_normalized.head()
```

```python
plt.figure(figsize=(18,14))
sns.heatmap(corr_spearman_normalized, cmap='coolwarm', square=True, cbar=True)
plt.title('Mapa de Calor de Correlación (Spearman) - Dataset Normalizado')
plt.show()
```

```python
# Selección de features para visualizar
features_vis = ['standard_deviation_e1', 'root_mean_square_e1', 'minimum_e1', 'maximum_e1']

# Pairplot con datos normalizados
sns.pairplot(df_normalized, vars=features_vis, hue='gesture', palette='muted')
plt.suptitle('Agrupación de Gestos en Espacio de Características (Normalizado)', y=1.02)
plt.show()
```

```python
import numpy as np

def calcular_gb_score(fila):
    score = 0
    
    # Umbrales aproximados en rango normalizado (ajustable)
    if fila['maximum_e1'] < 0.3: score += 1
    if fila['standard_deviation_e1'] < 0.3: score += 1
    if fila['root_mean_square_e1'] < 0.3: score += 1
    if fila['zero_crossings_e1'] < 0.3: score += 1
    if fila['wave_form_length_e1'] < 0.3: score += 1
    if fila['willison_amplitude_e1'] < 0.3: score += 1
    
    # Asignamos gb_score según cantidad de criterios
    if score >= 5:
        return 3  # ALTA probabilidad
    elif score >= 3:
        return 2  # MEDIA
    elif score >= 1:
        return 1  # BAJA
    else:
        return 0  # Ninguna

# Aplicamos la función al dataset
df_normalized['gb_score'] = df_normalized.apply(calcular_gb_score, axis=1)

# Verificamos
df_normalized.head()
```

```python
# Guardamos el dataset con gb_score
df_normalized.to_csv('emg_normalized_labeled_score.csv', index=False)

print("✅ Dataset guardado como 'emg_normalized_labeled_score.csv'")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Conteo de cada score
conteo_scores = df_normalized['gb_score'].value_counts().sort_index()

# Mostrar conteo
print(conteo_scores)

# Gráfico de barras
plt.figure(figsize=(6,4))
sns.barplot(x=conteo_scores.index, y=conteo_scores.values, palette='viridis')
plt.xlabel('gb_score')
plt.ylabel('Cantidad de muestras')
plt.title('Distribución de gb_score en el dataset')
plt.show()
```

```python
# Tomamos 5 muestras al azar de cada score
muestras_0 = df_normalized[df_normalized['gb_score'] == 0].sample(5, random_state=42)
muestras_1 = df_normalized[df_normalized['gb_score'] == 1].sample(5, random_state=42)
muestras_2 = df_normalized[df_normalized['gb_score'] == 2].sample(5, random_state=42)
muestras_3 = df_normalized[df_normalized['gb_score'] == 3].sample(5, random_state=42)
```

```python
# Columnas de análisis
cols_analisis = ['maximum_e1', 'standard_deviation_e1', 'root_mean_square_e1', 
                 'zero_crossings_e1', 'wave_form_length_e1', 'willison_amplitude_e1']

# Revisar muestras con score 0
print("🔍 Muestras con score 0")
print(muestras_0[cols_analisis])

# Revisar muestras con score 1
print("\n🔍 Muestras con score 1")
print(muestras_1[cols_analisis])

# Revisar muestras con score 2
print("\n🔍 Muestras con score 2")
print(muestras_2[cols_analisis])

# Revisar muestras con score 3
print("\n🔍 Muestras con score 3")
print(muestras_3[cols_analisis])
```

```python
import matplotlib.pyplot as plt

# Función para graficar los valores de cada muestra
def graficar_muestra(muestras, score):
    muestras[cols_analisis].T.plot(kind='bar', figsize=(10,6))
    plt.axhline(0.3, color='red', linestyle='--', label='Umbral 0.3')
    plt.title(f'Valores de características - Score {score}')
    plt.ylabel('Valor Normalizado')
    plt.xlabel('Características')
    plt.legend()
    plt.show()

# Graficamos cada grupo
graficar_muestra(muestras_0, 0)
graficar_muestra(muestras_1, 1)
graficar_muestra(muestras_2, 2)
graficar_muestra(muestras_3, 3)
```
