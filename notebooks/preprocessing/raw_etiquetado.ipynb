```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("emg_all_features_labeled.csv", keep_default_na=False)
df.head()
```

```python
df.columns
```

```python
# Definimos los nombres de los features
features = ['standard_deviation', 'root_mean_square', 'minimum', 'maximum',
            'zero_crossings', 'average_amplitude_change', 'amplitude_first_burst',
            'mean_absolute_value', 'wave_form_length', 'willison_amplitude']

# Ahora generamos las etiquetas para los 80 features (8 electrodos x 10 features)
column_names = []
for feature in features:
    for i in range(1, 9):  # 8 electrodos
        column_names.append(f'{feature}_e{i}')

# Agregamos la columna de etiqueta final
column_names.append('label')
```

```python
# Asignamos los nombres a las columnas
df.columns = column_names

# Verificamos el cambio
df.head()
```

```python
# Diccionario de etiquetas
label_map = {
    1: 'index_finger',
    2: 'middle_finger',
    3: 'ring_finger',
    4: 'little_finger',
    5: 'thumb',
    6: 'rest',
    7: 'victory_gesture'
}

# Creamos una nueva columna con el nombre de la etiqueta
df['gesture'] = df['label'].map(label_map)

# Mostramos las primeras filas con la nueva columna
df.head()
```

```python
# Guardamos el dataset etiquetado
df.to_csv('dataset_etiquetado.csv', index=False)
```
